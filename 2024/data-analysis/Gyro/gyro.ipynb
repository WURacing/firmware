```python
import pandas as pd
import numpy as np

# Read the CSV files and set the first row as the header
dfX = pd.read_csv('FGyroX.csv', header=0)
dfY = pd.read_csv('FGyroY.csv', header=0)
dfZ = pd.read_csv('FGyroZ.csv', header=0)



X_offset = dfX[' value'][0]

# Subtract the offset from the values

dfX[' value'] = dfX[' value'] - X_offset

Y_offset = dfY[' value'][0]

# Subtract the offset from the values

dfY[' value'] = dfY[' value'] - Y_offset

Z_offset = dfZ[' value'][0]

# Subtract the offset from the values

dfZ[' value'] = dfZ[' value'] - Z_offset



# Ensure the time columns are numeric
dfX['time'] = pd.to_numeric(dfX['time'])
dfY['time'] = pd.to_numeric(dfY['time'])
dfZ['time'] = pd.to_numeric(dfZ['time'])

# Integrate using the trapezoidal rule
def integrate(df, time_col, value_col):
    return np.trapz(df[value_col], df[time_col])

integral_X = integrate(dfX, 'time', ' value')
integral_Y = integrate(dfY, 'time', ' value')
integral_Z = integrate(dfZ, 'time', ' value')

print("Integral of FGyroX values over time:", integral_X)
print("Integral of FGyroY values over time:", integral_Y)
print("Integral of FGyroZ values over time:", integral_Z)




#plot all 3

import matplotlib.pyplot as plt

plt.plot(dfX['time'], dfX[' value'], label='FGyroX')

plt.plot(dfY['time'], dfY[' value'], label='FGyroY')

plt.plot(dfZ['time'], dfZ[' value'], label='FGyroZ')

plt.legend()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the CSV files and set the first row as the header
dfX = pd.read_csv('FGyroX.csv', header=0)
dfY = pd.read_csv('FGyroY.csv', header=0)
dfZ = pd.read_csv('FGyroZ.csv', header=0)

# Subtract the offset from the values
X_offset = dfX[' value'][0]
dfX[' value'] = dfX[' value'] - X_offset

Y_offset = dfY[' value'][0]
dfY[' value'] = dfY[' value'] - Y_offset

Z_offset = dfZ[' value'][0]
dfZ[' value'] = dfZ[' value'] - Z_offset

# Ensure the time columns are numeric
dfX['time'] = pd.to_numeric(dfX['time'])
dfY['time'] = pd.to_numeric(dfY['time'])
dfZ['time'] = pd.to_numeric(dfZ['time'])

# Integrate using the trapezoidal rule
def integrate(df, time_col, value_col):
    return np.trapz(df[value_col], df[time_col])

integral_X = integrate(dfX, 'time', ' value')
integral_Y = integrate(dfY, 'time', ' value')
integral_Z = integrate(dfZ, 'time', ' value')

print("Integral of FGyroX values over time:", integral_X)
print("Integral of FGyroY values over time:", integral_Y)
print("Integral of FGyroZ values over time:", integral_Z)

# Calculate the rolling average with a window size of 5 (adjust as needed)
dfX['rolling_avg'] = dfX[' value'].rolling(window=20).mean()
dfY['rolling_avg'] = dfY[' value'].rolling(window=20).mean()
dfZ['rolling_avg'] = dfZ[' value'].rolling(window=20).mean()
# Integrate the rolling averages
integral_X_rolling = integrate(dfX.dropna(), 'time', 'rolling_avg')
integral_Y_rolling = integrate(dfY.dropna(), 'time', 'rolling_avg')
integral_Z_rolling = integrate(dfZ.dropna(), 'time', 'rolling_avg')

print("Integral of FGyroX rolling average over time:", integral_X_rolling)
print("Integral of FGyroY rolling average over time:", integral_Y_rolling)
print("Integral of FGyroZ rolling average over time:", integral_Z_rolling)

# Plot all 3 original values and rolling averages
plt.figure(figsize=(12, 6))
plt.plot(dfX['time'], dfX[' value'], label='FGyroX')
plt.plot(dfX['time'], dfX['rolling_avg'], label='FGyroX Rolling Avg', linestyle='--')

plt.plot(dfY['time'], dfY[' value'], label='FGyroY')
plt.plot(dfY['time'], dfY['rolling_avg'], label='FGyroY Rolling Avg', linestyle='--')

plt.plot(dfZ['time'], dfZ[' value'], label='FGyroZ')
plt.plot(dfZ['time'], dfZ['rolling_avg'], label='FGyroZ Rolling Avg', linestyle='--')

plt.legend()
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('FGyro Values and Rolling Averages')
plt.show()
```
